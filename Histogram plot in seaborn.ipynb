```python
# What is a Histogram?

# A histogram shows the distribution of a numerical variable by dividing the data into bins (intervals) and counting how many values fall into each bin.
# It’s great for understanding frequency distribution.
```

```python
# Histogram in Seaborn

# In Seaborn, we use sns.histplot().

# 1. x / y → variable to plot
# sns.histplot(x="total_bill", data=tips)

# 2.bins → number of intervals
# sns.histplot(x="total_bill", data=tips, bins=20)

# 3.hue → separate by another variable (different colors)
# sns.histplot(x="total_bill", data=tips, hue="sex")

# 4.kde=True → add a smooth line (Kernel Density Estimate)

# sns.histplot(x="total_bill", data=tips, kde=True)

# 5.multiple → overlapping or stacked histograms

# sns.histplot(x="total_bill", data=tips, hue="sex", multiple="stack")
```
